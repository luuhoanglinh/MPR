#### get heading_mpr function

```python
def get_heading_mpr(heading_mpr):
    heading_mpr = heading_mpr.split(",")
    heading_mpr = heading_mpr[0]

    if "phân tích" in heading_mpr:
        heading_mpr = heading_mpr.split("phân tích")
        heading_mpr = "Phân tích" + heading_mpr[1]
    elif "tổng quan" in heading_mpr:
        heading_mpr = heading_mpr.split("tổng quan")
        heading_mpr = "Tổng quan" + heading_mpr[1]
    elif "chuyên sâu" in heading_mpr:
        heading_mpr = heading_mpr.split("chuyên sâu")
        heading_mpr = "Chuyên sâu" + heading_mpr[1]
    
    return heading_mpr
```

#### get tong_quan function

```python
def get_tong_quan(df1,i):
    df2 = [] # df storing infos of "diem_noi_bat"
    while i < len(df1):
        if "Một số điểm nổi bật" not in df1[i]:
            df2.append(df1[i][2:])
            i = i + 1
        else:
            break

    tong_quan = ""
    for j in range(0,len(df2)):
        tong_quan = tong_quan + "<li>" + df2[j] + "</li>"

    return tong_quan,i
```
